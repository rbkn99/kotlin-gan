```python
import pandas as pd
```

```python
base_dir = '../full_data/'
actions_df = pd.read_csv(base_dir + 'actions.csv', index_col=0)
rules_df = pd.read_csv(base_dir + 'rules.csv', index_col=0)
data_df = pd.read_csv(base_dir + 'data.csv')
```

```python
rare_rules = rules_df[rules_df['files_counter'] < 10]
rare_actions = actions_df[actions_df['files_counter'] < 10]
```

```python
data_df.shape
```

```python
rare_files = set()
rare_rules_id = set(rare_rules.index)
rare_actions_id = set(rare_actions.index)
for file_id, group in data_df.groupby('file_id'):
    if group['rule_id'].isin(rare_rules_id).any() or group['action_id'].isin(rare_actions_id).any():
        rare_files.add(file_id)
```

```python
rules_df = rules_df[~rules_df.index.isin(rare_rules_id)]['rule_name']
actions_df = actions_df[~actions_df.index.isin(rare_actions_id)]['action_name']
data_df = data_df[~data_df.file_id.isin(rare_files)]
```

```python
new_ids_rules = {old_id: new_id for new_id, old_id in enumerate(rules_df.index)}
new_ids_actions = {old_id: new_id for new_id, old_id in enumerate(actions_df.index)}
data_df['rule_id'] = data_df['rule_id'].map(new_ids_rules)
data_df['action_id'] = data_df['action_id'].map(new_ids_actions)
```

```python
rules_df.index = list(range(len(rules_df)))
actions_df.index = list(range(len(actions_df)))
```

```python
rules_df.to_csv('../data/rules.csv')
actions_df.to_csv('../data/actions.csv')
data_df.to_csv('../data/data.csv', index=0)
```
